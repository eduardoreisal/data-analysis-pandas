```python
ages = [20,25,30,35,40,45]
salaries = [60000, 80000, 100000, 120000, 140000, 200000]
plt.xticks(ages)
plt.yticks(salaries, labels=['60k', '80k', '100k', '120k', '140k', '200k'])
plt.plot(ages, salaries, color='olive', linewidth=3)
plt.title('Salaries', fontsize=20)
plt.xlabel('Age', fontsize=16)
plt.ylabel('Salary', fontsize=16, labelpad=10)
```

```python
ages = [20,25,30,35,40,45]
salaries = [60000, 80000, 100000, 120000, 140000, 200000]
plt.xticks(ages)
plt.yticks(salaries, labels=['60k', '80k', '100k', '120k', '140k', '200k'])
plt.plot(ages, salaries, color='olive', linewidth=3)
plt.title('Salaries', fontsize=20)
plt.xlabel('Age', fontsize=16)
plt.ylabel('Salary', fontsize=16, labelpad=10)
plt.xlim(20,45)
plt.ylim(8000, 120000)
```

```python
nums = [60, 50, 40, 30, 10]
nums1 = [x*2 for x in nums]
nums2 = [x // 2 for x in nums]
plt.figure(figsize=(8,8))
plt.plot(nums, color="teal", label="x")
plt.plot(nums1, color='olive', label="x square")
plt.plot(nums2, color="purple", label="x cubed")
plt.legend(loc="upper right")
```
