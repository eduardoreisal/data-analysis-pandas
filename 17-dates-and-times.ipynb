```python
ufos["posted"] = pd.to_datetime(ufos["posted"])
ufos.info()
```

```python
ufos["time_before_reported"] =(ufos["posted"] - ufos['date_time'])
longest_wait = ufos["time_before_reported"].sort_values(ascending=False)
longest_wait
```
